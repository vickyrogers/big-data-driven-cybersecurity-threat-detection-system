```python
# COMPLETE CYBERSECURITY ANALYSIS WITH VISUALIZATION
print("🚀 Running Complete Cybersecurity Analysis...")

# Install required packages
!pip install pyspark==3.5.0 pandas matplotlib seaborn -q

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans
from pyspark.ml import Pipeline
from datetime import datetime, timedelta
import random

# Create Spark Session
spark = SparkSession.builder \
    .appName("CybersecurityAnalysis") \
    .config("spark.sql.adaptive.enabled", "true") \
    .config("spark.driver.memory", "2g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("WARN")

try:
    print("📁 STEP 1: Loading your cybersecurity data...")

    # Load your uploaded file
    df = spark.read.option("header", "true").option("inferSchema", "true").csv('cybersecurity_threat_detection_logsa.csv')

    print(f"✅ Loaded {df.count()} rows, {len(df.columns)} columns")
    print("📋 Columns:", df.columns)

    # Show data sample
    print("\n📄 Data Sample:")
    df.show(5, truncate=False)

    print("\n🔧 STEP 2: Preprocessing data...")

    # Clean and prepare data
    processed_df = df \
        .na.fill({"bytes_transferred": 0}) \
        .na.fill({"protocol": "UNKNOWN"}) \
        .na.fill({"action": "UNKNOWN"}) \
        .na.fill({"user_agent": "Unknown"}) \
        .na.fill({"request_path": "/"}) \
        .withColumn("timestamp", when(col("timestamp").isNull(), current_timestamp()).otherwise(col("timestamp"))) \
        .withColumn("is_malicious",
                   when((col("threat_label") != "Normal") &
                        (col("threat_label").isNotNull()), True)
                   .otherwise(False)) \
        .withColumn("hour_of_day", hour(col("timestamp"))) \
        .withColumn("has_malicious_ua",
                   lower(col("user_agent")).rlike("sqlmap|nmap|metasploit|nikto|scanner|hack")) \
        .withColumn("has_suspicious_path",
                   lower(col("request_path")).rlike("admin|phpmyadmin|wp-admin|shell|config|backup"))

    print("✅ Data preprocessing completed")

    print("\n🛡️ STEP 3: Running threat detection algorithms...")

    # 1. Detect Brute Force Attacks
    print("🔑 Detecting brute force attacks...")
    brute_force = processed_df.filter(lower(col("action")).isin(["deny", "fail", "reject", "block"])) \
        .groupBy("source_ip") \
        .agg(
            count("timestamp").alias("failed_attempts"),
            countDistinct("dest_ip").alias("target_servers"),
            min("timestamp").alias("first_attempt"),
            max("timestamp").alias("last_attempt")
        ) \
        .filter(col("failed_attempts") >= 5) \
        .withColumn("threat_type", lit("Brute Force Attack")) \
        .withColumn("confidence",
                   when(col("failed_attempts") > 20, "HIGH")
                   .when(col("failed_attempts") > 10, "MEDIUM")
                   .otherwise("LOW")) \
        .withColumn("detection_time", current_timestamp())

    # 2. Detect Suspicious User Agents
    print("👤 Detecting suspicious user agents...")
    suspicious_ua = processed_df.filter(col("has_malicious_ua") == True) \
        .groupBy("source_ip") \
        .agg(
            count("user_agent").alias("suspicious_requests"),
            countDistinct("user_agent").alias("unique_suspicious_ua"),
            collect_set("user_agent").alias("detected_tools")
        ) \
        .filter(col("suspicious_requests") > 0) \
        .withColumn("threat_type", lit("Suspicious User Agent")) \
        .withColumn("confidence",
                   when(col("unique_suspicious_ua") > 3, "HIGH")
                   .when(col("unique_suspicious_ua") > 1, "MEDIUM")
                   .otherwise("LOW")) \
        .withColumn("detection_time", current_timestamp())

    # 3. Detect Data Exfiltration
    print("📤 Detecting data exfiltration...")
    data_exfiltration = processed_df.filter(col("bytes_transferred") > 0) \
        .groupBy("source_ip") \
        .agg(
            sum("bytes_transferred").alias("total_data_sent"),
            count("dest_ip").alias("connection_count"),
            max("bytes_transferred").alias("largest_transfer")
        ) \
        .filter(col("total_data_sent") > 50000000) \
        .withColumn("threat_type", lit("Data Exfiltration")) \
        .withColumn("confidence",
                   when(col("total_data_sent") > 100000000, "HIGH")
                   .when(col("total_data_sent") > 50000000, "MEDIUM")
                   .otherwise("LOW")) \
        .withColumn("detection_time", current_timestamp())

    # 4. Detect Web Attacks
    print("🌐 Detecting web attacks...")
    sql_patterns = ["union", "select", "1=1", "sleep(", "drop table", "insert into"]
    xss_patterns = ["<script>", "javascript:", "alert(", "onload="]
    sql_expr = "|".join(sql_patterns)
    xss_expr = "|".join(xss_patterns)

    web_attacks = processed_df.filter(
        lower(col("request_path")).rlike(sql_expr) |
        lower(col("request_path")).rlike(xss_expr) |
        lower(col("user_agent")).rlike(sql_expr)
    ) \
    .groupBy("source_ip") \
    .agg(
        count("request_path").alias("attack_attempts"),
        collect_set("request_path").alias("attack_paths")
    ) \
    .filter(col("attack_attempts") > 0) \
    .withColumn("threat_type", lit("Web Application Attack")) \
    .withColumn("confidence",
               when(col("attack_attempts") > 10, "HIGH")
               .when(col("attack_attempts") > 5, "MEDIUM")
               .otherwise("LOW")) \
    .withColumn("detection_time", current_timestamp())

    print("\n🔗 STEP 4: Combining all threat detections...")

    # Combine all threats
    all_threats = brute_force.select("source_ip", "threat_type", "confidence", "detection_time") \
        .union(suspicious_ua.select("source_ip", "threat_type", "confidence", "detection_time")) \
        .union(data_exfiltration.select("source_ip", "threat_type", "confidence", "detection_time")) \
        .union(web_attacks.select("source_ip", "threat_type", "confidence", "detection_time")) \
        .distinct()

    # Store as final_results for visualization
    final_results = all_threats.cache()

    threat_count = final_results.count()
    unique_ips = final_results.select("source_ip").distinct().count()

    print(f"✅ Analysis completed! Found {threat_count} threats from {unique_ips} unique IPs")

    print("\n📊 STEP 5: Creating comprehensive visualizations...")

    # Create multiple visualizations
    plt.figure(figsize=(15, 10))

    # 1. Threat Type Distribution
    plt.subplot(2, 2, 1)
    threat_counts = final_results.groupBy("threat_type").count().toPandas()
    bars1 = plt.bar(threat_counts['threat_type'], threat_counts['count'],
                    color=['#ff6b6b', '#4ecdc4', '#45b7d1', '#96ceb4'])
    plt.title('Threat Type Distribution', fontweight='bold')
    plt.xticks(rotation=45, ha='right')
    for bar in bars1:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height, f'{int(height)}',
                ha='center', va='bottom', fontweight='bold')

    # 2. Confidence Level Distribution
    plt.subplot(2, 2, 2)
    confidence_counts = final_results.groupBy("confidence").count().toPandas()
    colors = {'HIGH': '#ff6b6b', 'MEDIUM': '#feca57', 'LOW': '#48dbfb'}
    bars2 = plt.bar(confidence_counts['confidence'], confidence_counts['count'],
                    color=[colors.get(conf, 'gray') for conf in confidence_counts['confidence']])
    plt.title('Confidence Level Distribution', fontweight='bold')
    for bar in bars2:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height, f'{int(height)}',
                ha='center', va='bottom', fontweight='bold')

    # 3. Top Suspicious IPs
    plt.subplot(2, 2, 3)
    ip_threats = final_results.groupBy("source_ip").count().orderBy("count", ascending=False).limit(8).toPandas()
    if not ip_threats.empty:
        bars3 = plt.bar(ip_threats['source_ip'], ip_threats['count'], color='#ff9ff3')
        plt.title('Top Suspicious IPs', fontweight='bold')
        plt.xticks(rotation=45, ha='right')
        for bar in bars3:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height, f'{int(height)}',
                    ha='center', va='bottom', fontweight='bold')
    else:
        plt.text(0.5, 0.5, 'No IP data', ha='center', va='center', transform=plt.gca().transAxes)
        plt.title('Top Suspicious IPs', fontweight='bold')

    # 4. Threats Over Time
    plt.subplot(2, 2, 4)
    time_threats = final_results.groupBy(hour("detection_time").alias("hour")).count().orderBy("hour").toPandas()
    if not time_threats.empty:
        plt.plot(time_threats['hour'], time_threats['count'], marker='o', linewidth=2, color='#54a0ff')
        plt.title('Threats by Hour of Detection', fontweight='bold')
        plt.xlabel('Hour of Day')
    else:
        plt.text(0.5, 0.5, 'No time data', ha='center', va='center', transform=plt.gca().transAxes)
        plt.title('Threats by Hour of Detection', fontweight='bold')

    plt.tight_layout()
    plt.show()

    print("\n📋 STEP 6: Displaying threat details...")

    # Show the actual detected threats
    print("🔍 DETECTED THREATS:")
    final_results.show(20, truncate=False)

    # Summary statistics
    print("\n📈 SUMMARY STATISTICS:")
    print(f"• Total threats detected: {threat_count}")
    print(f"• Unique suspicious IPs: {unique_ips}")
    print(f"• Threat types found: {final_results.select('threat_type').distinct().count()}")

    # High confidence threats
    high_conf = final_results.filter(col("confidence") == "HIGH").count()
    if high_conf > 0:
        print(f"• 🔴 HIGH confidence threats: {high_conf}")
        final_results.filter(col("confidence") == "HIGH").select("source_ip", "threat_type").show(truncate=False)

    print("\n🎉 CYBERSECURITY ANALYSIS COMPLETED SUCCESSFULLY!")

except Exception as e:
    print(f"❌ Error during analysis: {e}")
    import traceback
    traceback.print_exc()

finally:
    # Cleanup
    if 'final_results' in locals():
        final_results.unpersist()
    spark.stop()
    print("🛑 Spark session stopped")
```

```python
# COMPLETE CYBERSECURITY ANALYSIS WITH VISUALIZATION (FIXED VERSION)
print("🚀 Running Complete Cybersecurity Analysis...")

# Install required packages
!pip install pyspark==3.5.0 pandas matplotlib seaborn -q

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans
from pyspark.ml import Pipeline
from datetime import datetime, timedelta
import random

# Create Spark Session
spark = SparkSession.builder \
    .appName("CybersecurityAnalysis") \
    .config("spark.sql.adaptive.enabled", "true") \
    .config("spark.driver.memory", "2g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("WARN")

try:
    print("📁 STEP 1: Loading your cybersecurity data...")

    # Load your uploaded file
    df = spark.read.option("header", "true").option("inferSchema", "true").csv('cybersecurity_threat_detection_logsa.csv')

    print(f"✅ Loaded {df.count()} rows, {len(df.columns)} columns")
    print("📋 Columns:", df.columns)

    # Show data sample
    print("\n📄 Data Sample:")
    df.show(5, truncate=False)

    print("\n🔧 STEP 2: Preprocessing data...")

    # Clean and prepare data
    processed_df = df \
        .na.fill({"bytes_transferred": 0}) \
        .na.fill({"protocol": "UNKNOWN"}) \
        .na.fill({"action": "UNKNOWN"}) \
        .na.fill({"user_agent": "Unknown"}) \
        .na.fill({"request_path": "/"}) \
        .withColumn("timestamp", when(col("timestamp").isNull(), current_timestamp()).otherwise(col("timestamp"))) \
        .withColumn("is_malicious",
                   when((col("threat_label") != "Normal") &
                        (col("threat_label").isNotNull()), True)
                   .otherwise(False)) \
        .withColumn("hour_of_day", hour(col("timestamp"))) \
        .withColumn("has_malicious_ua",
                   lower(col("user_agent")).rlike("sqlmap|nmap|metasploit|nikto|scanner|hack")) \
        .withColumn("has_suspicious_path",
                   lower(col("request_path")).rlike("admin|phpmyadmin|wp-admin|shell|config|backup"))

    print("✅ Data preprocessing completed")

    print("\n🛡️ STEP 3: Running threat detection algorithms...")

    # 1. Detect Brute Force Attacks
    print("🔑 Detecting brute force attacks...")
    brute_force = processed_df.filter(lower(col("action")).isin(["deny", "fail", "reject", "block"])) \
        .groupBy("source_ip") \
        .agg(
            count("timestamp").alias("failed_attempts"),
            countDistinct("dest_ip").alias("target_servers"),
            min("timestamp").alias("first_attempt"),
            max("timestamp").alias("last_attempt")
        ) \
        .filter(col("failed_attempts") >= 5) \
        .withColumn("threat_type", lit("Brute Force Attack")) \
        .withColumn("confidence",
                   when(col("failed_attempts") > 20, "HIGH")
                   .when(col("failed_attempts") > 10, "MEDIUM")
                   .otherwise("LOW")) \
        .withColumn("detection_time", current_timestamp())

    print(f"   Found {brute_force.count()} brute force threats")

    # 2. Detect Suspicious User Agents
    print("👤 Detecting suspicious user agents...")
    suspicious_ua = processed_df.filter(col("has_malicious_ua") == True) \
        .groupBy("source_ip") \
        .agg(
            count("user_agent").alias("suspicious_requests"),
            countDistinct("user_agent").alias("unique_suspicious_ua"),
            collect_set("user_agent").alias("detected_tools")
        ) \
        .filter(col("suspicious_requests") > 0) \
        .withColumn("threat_type", lit("Suspicious User Agent")) \
        .withColumn("confidence",
                   when(col("unique_suspicious_ua") > 3, "HIGH")
                   .when(col("unique_suspicious_ua") > 1, "MEDIUM")
                   .otherwise("LOW")) \
        .withColumn("detection_time", current_timestamp())

    print(f"   Found {suspicious_ua.count()} suspicious user agent threats")

    # 3. Detect Data Exfiltration
    print("📤 Detecting data exfiltration...")
    data_exfiltration = processed_df.filter(col("bytes_transferred") > 0) \
        .groupBy("source_ip") \
        .agg(
            sum("bytes_transferred").alias("total_data_sent"),
            count("dest_ip").alias("connection_count"),
            max("bytes_transferred").alias("largest_transfer")
        ) \
        .filter(col("total_data_sent") > 50000000) \
        .withColumn("threat_type", lit("Data Exfiltration")) \
        .withColumn("confidence",
                   when(col("total_data_sent") > 100000000, "HIGH")
                   .when(col("total_data_sent") > 50000000, "MEDIUM")
                   .otherwise("LOW")) \
        .withColumn("detection_time", current_timestamp())

    print(f"   Found {data_exfiltration.count()} data exfiltration threats")

    # 4. Detect Web Attacks (FIXED REGEX PATTERNS)
    print("🌐 Detecting web attacks...")

    # FIXED: Use properly escaped regex patterns
    sql_patterns = ["union", "select", "1=1", "sleep", "drop table", "insert into", "or 1=1", "waitfor delay"]
    xss_patterns = ["script", "javascript", "alert", "onload", "onerror"]

    # Create regex patterns without special characters that need escaping
    sql_expr = "|".join(sql_patterns)
    xss_expr = "|".join(xss_patterns)

    web_attacks = processed_df.filter(
        lower(col("request_path")).rlike(sql_expr) |
        lower(col("request_path")).rlike(xss_expr) |
        lower(col("user_agent")).rlike(sql_expr)
    ) \
    .groupBy("source_ip") \
    .agg(
        count("request_path").alias("attack_attempts"),
        collect_set("request_path").alias("attack_paths")
    ) \
    .filter(col("attack_attempts") > 0) \
    .withColumn("threat_type", lit("Web Application Attack")) \
    .withColumn("confidence",
               when(col("attack_attempts") > 10, "HIGH")
               .when(col("attack_attempts") > 5, "MEDIUM")
               .otherwise("LOW")) \
    .withColumn("detection_time", current_timestamp())

    print(f"   Found {web_attacks.count()} web attack threats")

    print("\n🔗 STEP 4: Combining all threat detections...")

    # Combine all threats
    all_threats = brute_force.select("source_ip", "threat_type", "confidence", "detection_time") \
        .union(suspicious_ua.select("source_ip", "threat_type", "confidence", "detection_time")) \
        .union(data_exfiltration.select("source_ip", "threat_type", "confidence", "detection_time")) \
        .union(web_attacks.select("source_ip", "threat_type", "confidence", "detection_time")) \
        .distinct()

    # Store as final_results for visualization
    final_results = all_threats.cache()

    threat_count = final_results.count()
    unique_ips = final_results.select("source_ip").distinct().count()

    print(f"✅ Analysis completed! Found {threat_count} threats from {unique_ips} unique IPs")

    print("\n📊 STEP 5: Creating comprehensive visualizations...")

    # Create multiple visualizations
    plt.figure(figsize=(15, 10))

    # 1. Threat Type Distribution
    plt.subplot(2, 2, 1)
    threat_counts = final_results.groupBy("threat_type").count().toPandas()
    if not threat_counts.empty:
        bars1 = plt.bar(threat_counts['threat_type'], threat_counts['count'],
                        color=['#ff6b6b', '#4ecdc4', '#45b7d1', '#96ceb4'])
        plt.title('Threat Type Distribution', fontweight='bold')
        plt.xticks(rotation=45, ha='right')
        for bar in bars1:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height, f'{int(height)}',
                    ha='center', va='bottom', fontweight='bold')
    else:
        plt.text(0.5, 0.5, 'No threats detected', ha='center', va='center', transform=plt.gca().transAxes)
        plt.title('Threat Type Distribution', fontweight='bold')

    # 2. Confidence Level Distribution
    plt.subplot(2, 2, 2)
    confidence_counts = final_results.groupBy("confidence").count().toPandas()
    if not confidence_counts.empty:
        colors = {'HIGH': '#ff6b6b', 'MEDIUM': '#feca57', 'LOW': '#48dbfb'}
        bars2 = plt.bar(confidence_counts['confidence'], confidence_counts['count'],
                        color=[colors.get(conf, 'gray') for conf in confidence_counts['confidence']])
        plt.title('Confidence Level Distribution', fontweight='bold')
        for bar in bars2:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height, f'{int(height)}',
                    ha='center', va='bottom', fontweight='bold')
    else:
        plt.text(0.5, 0.5, 'No confidence data', ha='center', va='center', transform=plt.gca().transAxes)
        plt.title('Confidence Level Distribution', fontweight='bold')

    # 3. Top Suspicious IPs
    plt.subplot(2, 2, 3)
    ip_threats = final_results.groupBy("source_ip").count().orderBy("count", ascending=False).limit(8).toPandas()
    if not ip_threats.empty:
        bars3 = plt.bar(ip_threats['source_ip'], ip_threats['count'], color='#ff9ff3')
        plt.title('Top Suspicious IPs', fontweight='bold')
        plt.xticks(rotation=45, ha='right')
        for bar in bars3:
            height = bar.get_height()
            plt.text(bar.get_x() + bar.get_width()/2., height, f'{int(height)}',
                    ha='center', va='bottom', fontweight='bold')
    else:
        plt.text(0.5, 0.5, 'No IP data', ha='center', va='center', transform=plt.gca().transAxes)
        plt.title('Top Suspicious IPs', fontweight='bold')

    # 4. Threats Over Time
    plt.subplot(2, 2, 4)
    time_threats = final_results.groupBy(hour("detection_time").alias("hour")).count().orderBy("hour").toPandas()
    if not time_threats.empty:
        plt.plot(time_threats['hour'], time_threats['count'], marker='o', linewidth=2, color='#54a0ff')
        plt.title('Threats by Hour of Detection', fontweight='bold')
        plt.xlabel('Hour of Day')
    else:
        plt.text(0.5, 0.5, 'No time data', ha='center', va='center', transform=plt.gca().transAxes)
        plt.title('Threats by Hour of Detection', fontweight='bold')

    plt.tight_layout()
    plt.show()

    print("\n📋 STEP 6: Displaying threat details...")

    # Show the actual detected threats
    print("🔍 DETECTED THREATS:")
    final_results.show(20, truncate=False)

    # Summary statistics
    print("\n📈 SUMMARY STATISTICS:")
    print(f"• Total threats detected: {threat_count}")
    print(f"• Unique suspicious IPs: {unique_ips}")
    print(f"• Threat types found: {final_results.select('threat_type').distinct().count()}")

    # High confidence threats
    high_conf = final_results.filter(col("confidence") == "HIGH").count()
    if high_conf > 0:
        print(f"• 🔴 HIGH confidence threats: {high_conf}")
        final_results.filter(col("confidence") == "HIGH").select("source_ip", "threat_type").show(truncate=False)

    print("\n🎉 CYBERSECURITY ANALYSIS COMPLETED SUCCESSFULLY!")

except Exception as e:
    print(f"❌ Error during analysis: {e}")
    import traceback
    traceback.print_exc()

finally:
    # Cleanup
    if 'final_results' in locals():
        final_results.unpersist()
    spark.stop()
    print("🛑 Spark session stopped")
```
